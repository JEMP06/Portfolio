```python
# Haciendo la función algoritmica desde cero para mediana y media de max tenemos:
def promedio_custom(serie):
    return serie.sum() / serie.count()

def mediana_custom(serie):
    import numpy as np

    serie_ordenada = serie.sort_values(ascending=True)
    length_of_series = serie.count()

    if length_of_series % 2 == 0:
        bottom =int(length_of_series / 2)
        top = int(bottom + 1)
        return (serie_ordenada.iloc[bottom] + serie_ordenada.iloc[top]) / 2
    else:
        index = int(np.ceil(length_of_series / 2))
        return serie_ordenada.iloc[index]

import pandas as pd
df_meteoritos = pd.read_csv('../content/drive/MyDrive/Code/BEDU/All_Datasets/BEDU_Datasets/near_earth_objects-jan_feb_1995-raw.csv', index_col=0)
promedio_diametro = promedio_custom(df_meteoritos['estimated_diameter.meters.estimated_diameter_max'])
mediana_diametro = mediana_custom(df_meteoritos['estimated_diameter.meters.estimated_diameter_max'])
```

```python
# Verificación de Resultados:
def verificar_resultados(df_meteoritos, promedio_diametro, mediana_diametro):
    promedio = df_meteoritos['estimated_diameter.meters.estimated_diameter_max'].mean()
    if round(promedio_diametro, 4) != round(promedio, 4):
        print(f'El promedio no fue calculado correctamente.')
        print(f'Promedio esperado: {promedio}; Promedio recibido: {promedio_diametro}')
        return


    mediana = df_meteoritos['estimated_diameter.meters.estimated_diameter_max'].median()
    if round(mediana_diametro, 4) != round(mediana, 4):
        print(f'La mediana no fue calculada correctamente.')
        print(f'Mediana esperada: {mediana}; Mediana recibida: {mediana_diametro}')
        return

    print(f'Los estimados fueron calcualados correctamente.\n')
    print(f'El promedio calculado fue de {promedio_diametro} mientras que la mediana fue de {mediana_diametro}.')
    print(f'¿A qué le atribuyes tú la diferencia?')
verificar_resultados(df_meteoritos, promedio_diametro, mediana_diametro)
```

```python
import pandas as pd

df = pd.read_csv('../content/drive/MyDrive/Code/BEDU/All_Datasets/BEDU_Datasets/near_earth_objects-jan_feb_1995-raw.csv', index_col=0)
diameter_column = 'estimated_diameter.meters.estimated_diameter_max'
total_count = df.shape[0]
mean = df[diameter_column].mean()
std = df[diameter_column].std()

within_one_std_filter_bottom = df[diameter_column] >= (mean - std)
within_one_std_filter_top = df[diameter_column] <= (mean + std)
within_one_std = df[within_one_std_filter_bottom & within_one_std_filter_top]
percentage_of_data_within_one_std = within_one_std.shape[0] * 100 / total_count

within_two_std_filter_bottom = df[diameter_column] >= (mean - 2 * std)
within_two_std_filter_top = df[diameter_column] <= (mean + 2 * std)
within_two_std = df[within_two_std_filter_bottom & within_two_std_filter_top]
percentage_of_data_within_two_std = within_two_std.shape[0] * 100 / total_count

within_three_std_filter_bottom = df[diameter_column] >= (mean - 3 * std)
within_three_std_filter_top = df[diameter_column] <= (mean + 3 * std)
within_three_std = df[within_three_std_filter_bottom & within_three_std_filter_top]
percentage_of_data_within_three_std = within_three_std.shape[0] * 100 / total_count
```

```python
import pandas as pd

def calcular_distancia(serie, distancia):
  desviacion = serie.std(ddof=0)
  promedio = serie.mean()

  inferior = (promedio - (desviacion * distancia))
  superior = (promedio + (desviacion * distancia))

  en_rango = ((serie >= inferior) == (serie <= superior))

  suma = en_rango.sum()
  cantidad = serie.count()
  porcentaje = suma / cantidad

  return porcentaje

df_meteoritos = pd.read_csv("../content/drive/MyDrive/Code/BEDU/All_Datasets/BEDU_Datasets/near_earth_objects-jan_feb_1995-raw.csv")
diametros = df_meteoritos['estimated_diameter.meters.estimated_diameter_max']
cantidad_total_muestras = diametros.count()

distancia_1 = calcular_distancia(diametros, 1)
distancia_2 = calcular_distancia(diametros, 2)
distancia_3 = calcular_distancia(diametros, 3)
print(f'distancia de 1 desviacion: {distancia_1*100}%')
print(f'distancia de 2 desviacion: {distancia_2*100}%')
print(f'distancia de 3 desviacion: {distancia_3*100}%')
```

NOTA: Hay una regla llamada Seis Sigma que refiere a las 3 desviaciones estandar del lado izquiero y otras 3 del lado derecho, si hay algún dato despues de esas 3 sigmas de cualquiera de los lados en una distribución normal o campana de Gauss, entonces se considera como un valor
