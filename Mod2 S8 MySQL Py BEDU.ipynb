```python
#### RETO 3 ####

# Paso 1: Importar las bibliotecas necesarias
import mysql.connector as mysqlc
import pandas as pd

try:
    # Paso 2: Definir los datos de conexión a la base de datos MySQL
    cnx = mysqlc.connect(
        host="ec2-34-220-220-25.us-west-2.compute.amazonaws.com",
        port=3306,
        user="data",
        password='khQ9--mp6LWq97iA',
        database='movielens'
    )

    cursor = cnx.cursor()

    cursor.execute("SHOW TABLES")

    result = cursor.fetchall()

    result

    # Paso 4: Leer los datos de la tabla 'users' y convertirlos en un DataFrame
    users = pd.read_sql("SELECT * FROM users", cnx)
    # Paso 5: Leer los datos de la tabla 'occupations' y convertirlos en un DataFrame
    occupations = pd.read_sql("SELECT * FROM occupations", cnx)
    # Paso 6: Leer los datos de la tabla 'age_ranges' y convertirlos en un DataFrame
    age_ranges = pd.read_sql("SELECT * FROM age_ranges", cnx)

    # Paso 7: Combinar la información de 'users' y 'occupations' mediante el método merge
    users = pd.merge(users, occupations, left_on='occupation', right_on='occupation_id')
    # Paso 8: Combinar la información de 'users' y 'age_ranges' mediante el método merge
    users = pd.merge(users, age_ranges, left_on='age', right_on='age_id')
    # Paso 9: Renombrar y reordenar columnas para mayor claridad
    users = users.rename(columns={'description': 'occupation_description', 'range': 'age_range'})
    # Paso 10: Guardar el nuevo DataFrame como un archivo CSV
    users.to_csv('users_merged.csv', index=False)
    # Paso 11: Verificar el resultado
    users.head()

    # Cierra el cursor y la conexión
    cursor.close()
    cnx.close()

except mysql.connector.Error as err:
    print(f"Error al conectar a la base de datos: {err}")
```
